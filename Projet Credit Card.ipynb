```python
df_train['Job title'] = df_train['Job title'].replace({None: 'Laborers'})
df_test['Job title'] = df_test['Job title'].replace({None: 'Laborers'})
```

```python
tonum= ['Gender','Has a car','Has a property','Employment status','Education level','Marital status','Dwelling', 'Job title']
 
```

```python

for col in tonum:
    df_train[col], hashed= make_numerics(df_train[col], space_requested= True)
    df_test[col]= df_test[col].replace(hashed)
```

```python
df_train.head()
```

```python
y_train = df_train['Is high risk'].values
del df_train['Is high risk']
X_train = df_train.values
del df_train

y_test = df_test['Is high risk'].values
del df_test['Is high risk']
X_test = df_test.values
del df_test
```

Model

```python
from wolta.model_tools import compare_models

compare_models('clf',
              ['cat', 'raf', 'ada', 'per', 'svc', 'log'],
              ['acc', 'precision'],
              X_train, y_train, X_test, y_test)
```
